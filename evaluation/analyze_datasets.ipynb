```python
import pandas as pd
from evaluate import load_runs_metadata_table
```

```python
df = load_runs_metadata_table()
```

```python
def add_qrel_stats(dataset, subcollection):
    qrels_path = "../data/qrels/{dataset}-{subcollection}.qrels-core_queries"
    if dataset == "longeval" and subcollection != "WT":
        qrels_path = "../data/qrels/{dataset}-{subcollection}.qrels-test-core_queries"
    if dataset == "tripclick-test-head":
        qrels_path = "../data/qrels/{dataset}-{subcollection}.qrels-test-head-dctr-core_queries"

    path = qrels_path.format(dataset=dataset, subcollection=subcollection)
    qrels = pd.read_csv(path, delimiter=" ", names=["qid", "0", "docid", "rel"], dtype={"qid": str, "docid": str, "rel": int})
    total  = len(qrels)
    zero = len(qrels[qrels["rel"] == 0])
    one = len(qrels[qrels["rel"] == 1])
    two = len(qrels[qrels["rel"] == 2])
    return total, zero, one, two
    
```

```python
df[["qrels_sum", "qrel_0", "qrel_1", "qrel_2"]] = df.apply(lambda row: add_qrel_stats(row["dataset"], row["subcollection"]), axis=1, result_type='expand')
```

```python
df.drop_duplicates(subset=["dataset", "subcollection"], inplace=True)
```

```python
df.sort_values(by=["dataset", "subcollection"], inplace=True)
```
